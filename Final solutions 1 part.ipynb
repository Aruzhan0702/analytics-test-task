```python
import pandas as pd

df = pd.read_excel('Данные для тестового задания (2).xlsx', sheet_name='Данные об аудитории')
df['date'] = pd.to_datetime(df['date'])
df.head()
```

## MAU

```python
mau = df['user_id'].nunique()
print('MAU =', mau)
```

## DAU

```python
dau = df.groupby('date')['user_id'].nunique().mean()
print('DAU =', round(dau))
```

## 3. Retention

```python
first_visit = df.groupby('user_id')['date'].min().reset_index()
users_1_nov = first_visit[first_visit['date'] == '2023-11-01']['user_id']

returned_next_day = df[
    (df['user_id'].isin(users_1_nov)) &
    (df['date'] == '2023-11-02')
]['user_id'].nunique()

retention_day_1 = returned_next_day / len(users_1_nov) * 100
print('Retention Day 1 =', round(retention_day_1, 1), '%')
```

## 5. Пользовательская конверсия

```python
users_with_ad_view = df.groupby('user_id')['view_adverts'].sum().gt(0).sum()
all_users = df['user_id'].nunique()

conversion_ad_view = users_with_ad_view / all_users * 100
print('Конверсия в просмотр объявления =', round(conversion_ad_view, 1), '%')
```
